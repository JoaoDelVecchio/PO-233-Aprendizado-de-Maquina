```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('C:/Users/delve/OneDrive/Eu/GitHub/PO-233-Aprendizado-de-Maquina/Pré-Processamento/Normalized_Data.csv', index_col=0)
data2 = pd.read_csv('C:/Users/delve/OneDrive/Eu/GitHub/PO-233-Aprendizado-de-Maquina/Pré-Processamento/Non_Normalized_Data.csv', index_col=0)
data
```

```python
df = data
X = df.drop(columns=['Performance'])
y = df['Performance']

y = np.where(y > 0, 1, 0)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Checking the shape of the data
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Define the model
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(64, activation='sigmoid'),
    Dropout(0.5),
    Dense(64, activation='sigmoid'),
    Dropout(0.5),
    Dense(64, activation='sigmoid'),
    Dropout(0.5),
    Dense(64, activation='sigmoid'),
    Dropout(0.5),
    Dense(64, activation='sigmoid'),
    Dropout(0.5),
    Dense(64, activation='sigmoid'),
    Dropout(0.5),
    Dense(64, activation='sigmoid'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
# binary_crossentrop
# Train the model
history = model.fit(X_train, y_train,
                    epochs=300,
                    batch_size=32,
                    validation_split=0.2,
                    verbose=2)

# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)

test_loss, test_acc
```
