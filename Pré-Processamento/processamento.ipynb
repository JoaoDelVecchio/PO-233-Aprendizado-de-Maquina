Al√©m disso, vamos tornar todos os valores da nossa tabela em num√©ricos. Caso n√£o sejam num√©ricos, estar√£o classificados como NaN

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Mostrar informa√ß√µes sobre valores ausentes
missing_data = df.isnull().sum()

# Exibir as primeiras linhas ap√≥s convers√£o
print(missing_data)
```
