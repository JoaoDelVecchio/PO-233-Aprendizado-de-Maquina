# Eliminação de colunas desnecessárias

Vamos então primeiro mostrar as colunas que possuem valores ausentes, para realizar o tratamento.


Além disso, vamos tornar todos os valores da nossa tabela em numéricos. Caso não sejam numéricos, estarão classificados como NaN

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Mostrar informações sobre valores ausentes
missing_data = df.isnull().sum()

# Exibir as primeiras linhas após conversão
print(missing_data)
```

```python
# Definir um limite de 10% de valores ausentes
# limite_ausentes = 0.1 * len(df)
missing_percentage = df.isnull().mean()

# Remover colunas com mais de 10% de valores ausentes
# df = df.dropna(thresh=limite_ausentes, axis=1)
cols_to_drop = missing_percentage[missing_percentage > 0.1].index

# Preencher os valores ausentes restantes com a mediana da coluna
# 
df_cleaned = df.drop(columns=cols_to_drop)


# Verificar o resultado
# df
df = df_cleaned
df = df.fillna(df.median())
df.isnull().sum().sum()  # Verificando se ainda há valores ausentes
df = df.drop(df.columns[0], axis = 1)
df.to_csv('Non_Normalized_Data.csv')
df
```
