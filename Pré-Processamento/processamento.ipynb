# Elimina√ß√£o de colunas desnecess√°rias

Vamos ent√£o primeiro mostrar as colunas que possuem valores ausentes, para realizar o tratamento.


Al√©m disso, vamos tornar todos os valores da nossa tabela em num√©ricos. Caso n√£o sejam num√©ricos, estar√£o classificados como NaN

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Mostrar informa√ß√µes sobre valores ausentes
missing_data = df.isnull().sum()

# Exibir as primeiras linhas ap√≥s convers√£o
missing_data
#len(df.columns)
```

```python
# Definir um limite de 10% de valores ausentes
# limite_ausentes = 0.1 * len(df)
missing_percentage = df.isnull().mean()

# Remover colunas com mais de 10% de valores ausentes
# df = df.dropna(thresh=limite_ausentes, axis=1)
cols_to_drop = missing_percentage[missing_percentage > 0.1].index

# Preencher os valores ausentes restantes com a mediana da coluna
# 
df_cleaned = df.drop(columns=cols_to_drop)


# Verificar o resultado
# df
df = df_cleaned
df = df.fillna(df.median())
df.isnull().sum().sum()  # Verificando se ainda h√° valores ausentes
df = df.drop(df.columns[0], axis = 1)
df.to_csv('Non_Normalized_Data.csv')
df
```

# Elimina√ß√£o de Outliers

* Antes de normalizar meus valores, vamos remover os outliers, para isso, primeiro vamos calcular o z-score de cada elemento, isso √©, o valor de cada elemento caso a coluna fosse normalizada entre -1 e 1. Com o z-score calculado, vamos escolher um limite para o qual escolheremos os valores nessa nosva distruibui√ß√£o entre -1 e 1


```python
import numpy as np
data = df
# Calcular z-scores
data_zscores = (data - data.mean()) / data.std()

# Definir limiar de z-score
zscore_threshold = 5

# Filtrar dados
clean_data_zscore = data[(np.abs(data_zscores) < zscore_threshold).all(axis=1)]

# Exibir as primeiras linhas dos dados limpos
data = clean_data_zscore
len(data)
data
```

* Antes da elimina√ß√£o, estavamos com 2533 linhas, ap√≥s ficamos com apenas 2272
