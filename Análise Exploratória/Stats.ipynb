```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular a matriz de correlação
correlation_matrix = data.corr()

# Selecionar a coluna 'Performance' para ver as correlações específicas
performance_correlation = correlation_matrix['Performance'].sort_values(ascending=False)

# Visualizar as 10 variáveis mais correlacionadas
top_10_correlated = performance_correlation  # incluindo 'Performance' em si

# Plotar as correlações
plt.figure(figsize=(10, 8))
sns.barplot(x=top_10_correlated.index, y=top_10_correlated.values)
plt.title('Top 10 Variáveis Mais Correlacionadas com a Performance Relativa')
plt.xlabel('Variáveis')
plt.ylabel('Correlação de Pearson')
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

# Definir as variáveis independentes (X) e a variável dependente (y)
X = data.drop(columns=['Performance'])
y = data['Performance']

# Adicionar uma constante para o intercepto
X = sm.add_constant(X)

# Ajustar o modelo de regressão linear
model = sm.OLS(y, X).fit()

# Resumo do modelo
print(model.summary())
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import cross_val_score

# Regressão Ridge
ridge = Ridge(alpha=1.0)
ridge_scores = cross_val_score(ridge, X, y, cv=5)
print(f'Ridge Regression Scores: {ridge_scores.mean()}')

# Regressão Lasso
lasso = Lasso(alpha=0.1)
lasso_scores = cross_val_score(lasso, X, y, cv=5)
print(f'Lasso Regression Scores: {lasso_scores.mean()}')
```

```python
from sklearn.ensemble import RandomForestRegressor

# Ajustar o modelo Random Forest
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X, y)

# Obter a importância das variáveis
importances = rf.feature_importances_
feature_names = X.columns
feature_importances = pd.Series(importances, index=feature_names).sort_values(ascending=False)

print(feature_importances.head(10))
```

```python
from sklearn.feature_selection import RFE

# Utilizar Regressão Linear como estimador
estimator = LinearRegression()
selector = RFE(estimator, n_features_to_select=10, step=1)
selector = selector.fit(X, y)

# Mostrar as variáveis selecionadas
selected_features = X.columns[selector.support_]
selected_features
```

```python
from sklearn.model_selection import cross_val_score

# Utilizar Regressão Linear com validação cruzada
cv_scores = cross_val_score(LinearRegression(), X, y, cv=5)
print(f'Cross-Validation Scores: {cv_scores.mean()}')
```
